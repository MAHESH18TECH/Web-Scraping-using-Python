```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
driver = webdriver.Chrome("chromedriver")
Name_Lok=[]
driver.get("https://loksabhaph.nic.in/Members/Memberhomepage.aspx")
content = driver.page_source
soup = BeautifulSoup(content)
names =soup.findAll("tr" , class_="even")
names_2 =soup.findAll("tr" , class_="odd")
for data in names:
    for name in names_2:
        Name_element = data.find("a")
        Name_element_2 = name.find("a")
        print(Name_element.text)
        print(Name_element_2.text)
        Name_Lok.append(Name_element.text)
        Name_Lok.append(Name_element_2.text)
```
